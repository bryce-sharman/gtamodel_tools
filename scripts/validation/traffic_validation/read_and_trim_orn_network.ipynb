# Read and Trip Ontario Road Network (ORN) for Traffic Validation

Reads and trims the Ontario Road Network by: 1. Jurisdiction and 2. Road type to reduce file size and speedup file reading during model validation.

ORN can be downloaded from https://geohub.lio.gov.on.ca/datasets/mnrf::ontario-road-network-orn-road-net-element/explore

```python
from pathlib import Path
import gtamodel_tools.common.orn as orn
```

# Inputs

```python
validation_root = Path(r'S:\GTA Validation Data')
fp = validation_root / 'Traffic' / 'ORNELEM' / 'Non_Sensitive.gdb'
output_fp = validation_root / 'Traffic' / 'orn_gtha.gpkg'
```

```python
gdf = orn.read_geometry(fp)
jd = orn.read_jurisdiction(fp)
rc = orn.read_roadclass(fp)
```

```python
# The filter by jurisdiction filters elements based on their
# jurisdiction. It has three options 'GGH', or 'TTS_2022'
# Using 'GTHA' as this is for a GTAModel validation.
gdf_gtha = orn.filter_by_jurisdiction(gdf, jd, 'GTHA')
```

```python
# Remove laneways, trails, etc.
gdf_gtha2 = orn.filter_by_roadclass(
    gdf_gtha, rc, ['freeway', 'arterial', 'collector', 'local', 'transit'])
```

```python
gdf_gtha2.to_file(output_fp)
```
