# Prepare Toronto Centreline File 

Pre-process Toronto Centreline feospatial file to include cartesian directions for each line string based on geometry.

```python
from pathlib import Path

from gtamodel_tools.common.gis import prepare_gdf_for_count_processing
from gtamodel_tools.common.tcl import read_tcl
```

# Inputs

```python
root_dir = Path(r'S:\GTA Validation Data')
input_fp = root_dir / 'Centreline - Version 2 - 4326.gpkg'
output_fp = root_dir / 'Centreline - Version 2 - 4326 - Preprocessed.gpkg'

axis_offset = 17.0
```

```python
# Read Toronto TCL file
tcl_gdf = read_tcl(input_fp, False)
```

```python
# Preprocess this file to add direction information
prepared_tcl_gdf = prepare_gdf_for_count_processing(tcl_gdf, axis_offset)
```

```python
# Done, write to file, which will be read in directly
prepared_tcl_gdf.to_file(output_fp)
```
