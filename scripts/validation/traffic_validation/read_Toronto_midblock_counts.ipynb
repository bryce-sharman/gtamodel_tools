# Read and pre-process City of Toronto Mid-block Speed-Volume Classification (SVC) traffic counts

Reads City of Toronto SCV traffic counts, saving into traffic count database format expected by traffic validation tools. 

Note that the City of Toronto SVC traffic counts reference the Toronto Centreline shapefiles, which must also be downloaded.

Toronto SVC traffic counts can be found on Toronto Open Data at
https://open.toronto.ca/dataset/traffic-volumes-midblock-vehicle-speed-volume-and-classification-counts/

Toronto CentreLine shapefile can be found on Toronto Open Data at https://open.toronto.ca/dataset/toronto-centreline-tcl/

```python
from pathlib import Path

import gtamodel_tools.common.tcl as gis_tcl
from gtamodel_tools.validation.preprocess_traffic_counts.toronto_midblock_counts import read_midblock_counts
```

```python
# Inputs

validation_root = Path(r"S:\GTA Validation Data")
# Toronto CentreLine File
fp_tcl = validation_root / "Centreline - Version 2 - 4326.gpkg"
# Toronto traffic counts have been downloaded to the following directory
tcl_mdblk_root_dir = validation_root / "Toronto-Midblock-Counts-Nov2025"

# Output files
out_cnts_fp = validation_root / "Traffic" / "Toronto_midblock_counts.csv"
out_stns_fp = validation_root / "Traffic" / "Toronto_micblock_count_stns.gpkg"
```

```python
# Read in the Toronto Centreline Network
tcl_gdf = gis_tcl.read_tcl_gpkg(fp_tcl)
```

```python
# Read all downloaded Toronto mid-block count data
stns, cnts = read_midblock_counts(tcl_mdblk_root_dir, tcl_gdf)
```

```python
# Output to file
cnts.to_csv(out_cnts_fp)
stns.to_file(out_stns_fp)
```
